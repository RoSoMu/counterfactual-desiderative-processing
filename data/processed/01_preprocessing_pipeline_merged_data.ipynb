Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read data

```python
metadata = pd.read_csv("../../data/raw/metadata_raw.csv")
responses = pd.read_csv("../../data/raw/counterfactual_responses_raw.csv")
```

Responses dataset:

Cleaning leading or trailing whitespaces in "Response"

```python
responses["Response"] = responses["Response"].str.strip()
responses["Response"].value_counts(dropna= False)
```

```python
responses["Response"].shape
```

Rename "Participant_ID"

```python
responses = responses.rename(
    columns={"Partcipant_ID": "Participant_ID"}
)
```

Include Null values in "Response" as category

```python
responses["Response_Full"] = (
    responses["Response"]
    .fillna("null")
)
```

```python
responses["Response_Full"].value_counts()
```

Metadata dataset:

Clean leading and trailing whitespaces from columns

```python
metadata.columns = metadata.columns.str.strip()
```

Eliminate "Agent" column -included in responses

```python
metadata = metadata.drop(columns = ["Agent"])
```

Add know_level and per_level labels

```python
level_mapping = {
    1: "A1",
    2: "A2.1",
    3: "A2.2",
    4: "B1.1",
    5: "B1.2",
    6: "B2.1",
    7: "B2.2",
    8: "C1",
    9: "C2"
}

metadata["Kno_Level_Label"] = metadata["Kno_Level"].map(level_mapping)
metadata["Per_Level_Label"] = metadata["Per_Level"].map(level_mapping)
```

```python
level_order = [
    "A1",
    "A2.1",
    "A2.2",
    "B1.1",
    "B1.2",
    "B2.1",
    "B2.2",
    "C1",
    "C2"
]

metadata["Kno_Level_Label"] = pd.Categorical(
    metadata["Kno_Level_Label"],
    categories=level_order,
    ordered=True
)

metadata["Per_Level_Label"] = pd.Categorical(
    metadata["Per_Level_Label"],
    categories=level_order,
    ordered=True
)
```

Add prof_use and prof_need labels

```python
prof_mapping = {
    0: "No",
    1: "Yes"
}

metadata["Prof_Need_Label"] = (
    metadata["Prof_Need"]
    .map(prof_mapping)
)

metadata["Prof_Use_Label"] = (
    metadata["Prof_Use"]
    .map(prof_mapping)
)
```

Merge datasets

```python
merged_data = responses.merge(
    metadata,
    on= "Participant_ID",
    how="left"
)
```
