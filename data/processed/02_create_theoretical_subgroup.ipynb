Creates a theoretical subgroup by adding three more variables at an item level to see further information in the consequent: consequent_reversibility, goal_oriented_behaviour and emotional_load.

import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read data and target subgroup by Per_Level and Prof_Use

```python
merged_data = pd.read_csv("merged_clean.csv")

subgroup = merged_data[
    (merged_data["Per_Level_Label"].isin(["B2.1", "B2.2", "C1"])) &
    (merged_data["Prof_Use_Label"] == "Yes")
].copy()
```

```python
subgroup.shape
```

Lookup table: adds new columns and values

```python
item_features = pd.DataFrame({
    "Goal_Type": [
        "no_goal", "no_goal",
        "goal_non_frequent", "goal_non_frequent",
        "goal_frequent", "goal_frequent"
    ],
    "Item": ["Item1", "Item2", "Item1", "Item2", "Item1", "Item2"],
    "Consequent_Reversibility": ["Yes", "Yes", "No", "No", "Yes", "Yes"],
    "Goal_Oriented_Behaviour": ["No", "No", "No", "No", "Yes", "Yes"],
    "Emotional_Load": ["No", "No", "Yes", "Yes", "No", "No"]
})
```

Merge tables

```python
subgroup_theoretical = subgroup.merge(
    item_features,
    on=["Goal_Type", "Item"],
    how="left"
)
```

```python
subgroup_theoretical.shape
```

```python
subgroup_theoretical[
    ["Goal_Type", "Item", "Consequent_Reversibility",
     "Goal_Oriented_Behaviour", "Emotional_Load"]
].drop_duplicates()
```

Save new processed dataset

```python
subgroup_theoretical.to_csv("subgroup_theoretical.csv", index=False)
```
