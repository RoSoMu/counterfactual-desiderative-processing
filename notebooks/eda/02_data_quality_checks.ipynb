4. Counts per column

```python
for col in ["Focus", "Agent", "Goal_Type", "Item", "Response"]:
    print(col)
    print(responses[col].value_counts(dropna=False))
    print()
```

```python
for col in ["EXP", "Gender", "Kno_Level", "Per_Level", "Motivation", "GN_Score", "Difficulty"]:
    print(col)
    print(metadata[col].value_counts(dropna=False).sort_index())
    print()
```

5. Data cleaning

There seems to be an extra "L1_transfer" category.

```python
responses["Response"].unique()
```

Remove leading/ trailing spaces and counts check.

```python
responses["Response"] = responses["Response"].str.strip()
print(responses["Response"].value_counts(dropna=False))
```

Rename "Participant_ID" and check counts grouped by ID.

```python
responses = responses.rename(
    columns={"Partcipant_ID": "Participant_ID"}
)
```
