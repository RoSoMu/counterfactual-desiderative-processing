Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read data and sanity checks

```python
subgroup_theoretical = pd.read_csv("../../data/processed/subgroup_theoretical.csv")
```

```python
subgroup_theoretical[
    [
        "Goal_Type",
        "Item",
        "Consequent_Reversibility",
        "Goal_Oriented_Behaviour",
        "Emotional_Load"
    ]
].drop_duplicates()
```

```python
subgroup_theoretical.columns
```

```python
subgroup_theoretical.shape
```

Overall subgroup proportions

Response Proportion by Agent and Consequent Reversiblity

```python
agent_reversibility_props= pd.crosstab(
    [subgroup_theoretical["Agent"], subgroup_theoretical["Consequent_Reversibility"]],
    subgroup_theoretical["Response_Full"],
    normalize="index"
)

agent_reversibility_props
```

```python
agent_reversibility_props.plot(kind = "bar", stacked=True)

plt.title("Response Proportions by Agent and Consequent Reversibility")
plt.legend(title="Response", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xlabel("Consequent Reversibilty")
plt.ylabel("Proportion")
plt.show()
```
