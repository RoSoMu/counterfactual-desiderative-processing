# L2 Escape Model

Consequent Reversibility was not included in the final escape models because, in this subgroup structure, it is partially encoded by Goal_Type and produced unstable estimates when entered together with Goal_Type.

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Overall Subgroup:
1. Create escape_L2
2. Counts and Proportions
3. Logistic Models
4. Model Comparison

Read data

```python
subgroup_theoretical = pd.read_csv("../../data/processed/subgroup_theoretical.csv")
```

1. Create escape_2

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

2. Counts and Proportions

```python
escape_L2_counts = pd.crosstab(
    subgroup_theoretical["escape_L2"],
    subgroup_theoretical["Response_Full"])

escape_L2_counts
```

```python
escape_L2_props = subgroup_theoretical["escape_L2"].value_counts(normalize = True)
escape_L2_props
```

Counts and Proportions by Condition

```python
counts_condition = pd.crosstab(
    [subgroup_theoretical["Goal_Type"], subgroup_theoretical["Agent"], subgroup_theoretical["Consequent_Reversibility"]],
    subgroup_theoretical["escape_L2"],
    margins = True)

counts_condition
```

```python
props_condition = pd.crosstab(
    [subgroup_theoretical["Goal_Type"], subgroup_theoretical["Agent"], subgroup_theoretical["Consequent_Reversibility"]],
    subgroup_theoretical["escape_L2"],
    normalize= "index")

props_condition
```

Agency appears to help participants escape L2. 

The condition with the strongest L1 transfer (goal_non_frequent, agent=1) also exhibits the highest L2 escape rate (95.8%). This suggests that failure does not occur at the L2 suppression stage, but at the subsequent resolution stage.

In addition, if there is a goal representation available, agency seems to help participants commit by leaving the L2 space; however, if there is no goal representation, agency seems to have much less to work with. 

3. Logistic Models

Import Libraries

```python
import statsmodels.formula.api as smf
from scipy.stats import chi2
```

Binary Outcome

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

```python
subgroup_theoretical["escape_L2"].value_counts()
```

**Model E1**
escape_L2 ~ Goal_Type

```python
E1 = smf.logit(
    "escape_L2 ~ Goal_Type",
    data=subgroup_theoretical
    ).fit()

print(E1.summary())
```

**Model E2**
escape_L2 ~ Goal_Type + Agency

```python
E2_2 = smf.logit(
    "escape_L2 ~ Consequent_Reversibility + Agent",
    data= subgroup_theoretical
    ).fit()

print(E2_2.summary())
```

```python
E2 = smf.logit(
    "escape_L2 ~ Goal_Type + Agent",
    data= subgroup_theoretical
    ).fit()

print(E2.summary())
```

**Model E3**
escape_L2 ~ Consequent_Reversibility + Goal_Type + Agent

```python
E3 = smf.logit(
    "escape_L2 ~ Goal_Type + Agent",
    data=subgroup_theoretical
    ).fit()

print(E3.summary())
```

**Model E4**
escape_L2 ~ Goal_Type x Agent

```python
E4 = smf.logit(
    "escape_L2 ~ Goal_Type * Agent",
    data=subgroup_theoretical
).fit()

print(E4.summary())
```

```python
E4_2 = smf.logit(
    "escape_L2 ~ Consequent_Reversibility * Agent",
    data=subgroup_theoretical
).fit()

print(E4_2.summary())
```

4. Model Comparisons

```python
from scipy.stats import chi2

def lr_test(model_small, model_large):

    LR = 2 * (model_large.llf - model_small.llf)

    df = model_large.df_model - model_small.df_model

    p = chi2.sf(LR, df)

    return pd.Series({
        "LR": LR,
        "df": df,
        "p_value": p
    })
```

```python
lr_test(E3, E4)
```

4. Model comparisons

**E1 Model**: 
escape_L2 ~ Consequent_Reversibility

```python
E1 = smf.logit(
    "escape_L2 ~ Goal_Type",
    data=subgroup_actors
    ).fit()

print(E1.summary())
```

For actors, consequent reversibility did not significantly predict L2 escape (p = .629), suggesting that reversibility alone is insufficient to trigger abandonment of L2 interpretations. This supports the view that reversibility functions as a source of latent cognitive load whose effects become observable only when additional representational demands are introduced.
