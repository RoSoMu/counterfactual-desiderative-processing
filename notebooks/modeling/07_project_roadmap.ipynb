# Project Roadmap: Counterfactual-Desiderative Processing

## 1. Purpose of this notebook
This notebook explains the logic behind the modelling notebooks that follow. It documents the analytical decisions made after the exploratory phase and provides a map for readers.

The purpose of this project is not to identify isolated statistically significant predictors, but to investigate the stability of behavioural relationships under increasingly restrictive views of the data. Rather than treating data pruning as a nuisance, this project uses progressive pruning as a methodological tool to explore which behavioural effects persist, strengthen, weaken, or disappear across nested subsets of observations.

## 2. Main research pathway
Stage 1: Escape from L2  
Stage 2: Resolution after escape  
- resolved_correct
- resolved_transfer

## 3. Subgroups
### Overall subgroup
Participants meeting the proficiency/use criteria.

### Actors subgroup
Subset where the consequent begins with an actor/agentive structure.

### Why not readers vs actors?
Because the available observations were too limited once split further, and the central contrast became more interpretable as overall vs actors.

## 4. Why escape and resolution are modelled separately
Escape_L2 captures whether participants move away from L2_other.  
Resolution captures what they do after escaping: correct response vs L1_transfer.

## 5. Why Consequent_Reversibility was dropped
Consequent_Reversibility was explored, sometimes alongside Agent, but it did not provide a clearer theoretical or statistical contribution than Agent/Goal_Type. It also overlapped conceptually with the goal/agent structure, so it was removed from the main modelling pathway.
