Read Data

```python
subgroup_theoretical = pd.read_csv("../../data/processed/subgroup_theoretical.csv")
```

1. Create escape_2

Binary Outcome: responses that are not "L2_other" -those who managed to escape L2_other by choosing either a correct response or an L1_transfer response.

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

2. Counts and Proportions

```python
escape_L2_counts = pd.crosstab(
    subgroup_theoretical["escape_L2"],
    subgroup_theoretical["Response_Full"],
    margins = True)

escape_L2_counts
```

```python
escape_L2_props = subgroup_theoretical["escape_L2"].value_counts(normalize = True)
escape_L2_props
```

From 198 observations, almost 80% of responses managed to escape L2_other options. 

Counts and Proportions by Condition:

```python
counts_condition = pd.crosstab(
    [subgroup_theoretical["Goal_Type"], subgroup_theoretical["Agent"], subgroup_theoretical["Focus"]],
    subgroup_theoretical["escape_L2"],
    margins = True)

counts_condition
```

```python
props_condition = pd.crosstab(
    [subgroup_theoretical["Goal_Type"], subgroup_theoretical["Agent"], subgroup_theoretical["Focus"]],
    subgroup_theoretical["escape_L2"],
    normalize= "index")

props_condition
```

3. Logistic Models

Import Libraries

```python
import statsmodels.formula.api as smf
from scipy.stats import chi2
```

```python
subgroup_theoretical["escape_L2"].value_counts()
```

**Main Effect Models**

**E01**

escape_L2 ~ Focus

```python
E01 = smf.logit(
    "escape_L2 ~ Focus",
    data=subgroup_theoretical
    ).fit()

print(E01.summary())
```

Focus on its own cannot predict anything.

**E02**

escape_L2 ~ Agent

```python
E2 = smf.logit(
    "escape_L2 ~ Agent",
    data= subgroup_theoretical
    ).fit()

print(E2.summary())
```

Agency alone is strong.

**E03**

escape_L2 ~ Goal_Type

```python
E03 = smf.logit(
    "escape_L2 ~ Goal_Type",
    data= subgroup_theoretical
    ).fit()

print(E03.summary())
```

Goal Type on its alone predict anything.

**Additive Models**

**E04**

escape_L2 ~ Focus + Agent

```python
E04 = smf.logit(
    "escape_L2 ~ Focus + Agent",
    data=subgroup_theoretical
    ).fit()

print(E04.summary())
```

Focus is not individually significant; however, adding it to Agency produces a good model. Thus, Focus does not predict escaping L2 options on its own but accounting for it does not erase the agency effect. 

**E05**

escape_L2 ~ Focus + Goal_Type

```python
E05 = smf.logit(
    "escape_L2 ~ Focus + Goal_Type",
    data=subgroup_theoretical
    ).fit()

print(E05.summary())
```

As expected, neither predictor carries the model. 

**E06**

escape_L2 ~ Agent + Goal_Type

```python
E06 = smf.logit(
    "escape_L2 ~ Agent + Goal_Type",
    data=subgroup_theoretical
    ).fit()

print(E06.summary())
```

Interestingly, Goal Type so far did nothing on its own and nor when added to Focus; however, the additive model with Agency still works. 

```python
P01 = smf.logit(
    "escape_L2 ~ Focus + Agent + Goal_Type",
    data=subgroup_theoretical
    ).fit()

print(P01.summary())
```

**Interactive Models**

**E07**

escape_L2 ~ Focus * Agent

```python
E07 = smf.logit(
    "escape_L2 ~ Focus * Agent",
    data=subgroup_theoretical
    ).fit()

print(E07.summary())
```

Interaction Focus-Agency (They:Agent, p = 0.984) shows basically zero evidence for interaction. That simplifies interpretation. 

**E08**

escape_L2 ~ Focus * Goal_Type

```python
E08 = smf.logit(
    "escape_L2 ~ Focus * Goal_Type",
    data=subgroup_theoretical
    ).fit()

print(E08.summary())
```

Interaction Focus-Goal_Type non-frequent (They:goal_non_frequent) shows a large coefficient and a very high standard error. There could be something there but there is not enough precision. 

**E09**

escape_L2 ~ Agent * Goal_Type

```python
E09 = smf.logit(
    "escape_L2 ~ Agent * Goal_Type",
    data=subgroup_theoretical
    ).fit()

print(E09.summary())
```

Interaction Agent-Goal_Type still survives. 

**E10**

escape_L2 ~ Focus * Agent + Goal_Type

```python
E10 = smf.logit(
    "escape_L2 ~ Focus * Agent + Goal_Type",
    data=subgroup_theoretical
    ).fit()

print(E10.summary())
```

In the full additive with interaction model, Agency survives, Focus doesn't, Goal Type stays weak and interaction disappears. So, while it doesn't really work, consistency is valuable.

4. Model Comparisons

```python
from scipy.stats import chi2

def lr_test(model_small, model_large):

    LR = 2 * (model_large.llf - model_small.llf)

    df = model_large.df_model - model_small.df_model

    p = chi2.sf(LR, df)

    return pd.Series({
        "LR": LR,
        "df": df,
        "p_value": p
    })
```

**Behavioural hierarchy**

Focus (E01)

↓

Focus + Agent (E04)

```python
lr_test(E01, E04)
```

Focus (E01)

↓

Focus + Agent + Goal_Type (P01)

```python
lr_test(E01, P01)
```

Focus + Agent (E04)

↓

Focus + Agent + Goal_Type (P01)

```python
lr_test(E04, P01)
```

**Interaction hierarchy**

Focus + Agent (E04)

↓

Focus * Agent (E07)

```python
lr_test(E04, E07)
```

Focus + Goal_Type (E05)

↓

Focus * Goal_Type (E08)

```python
lr_test(E05, E08)
```

Agent + Goal_Type (E06)

↓

Agent * Goal_Type (E09)

```python
lr_test(E06, E09)
```
