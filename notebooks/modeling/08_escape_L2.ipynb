# L2 Escape Model

Consequent Reversibility was not included in the final escape models because, in this subgroup structure, it is partially encoded by Goal_Type and produced unstable estimates when entered together with Goal_Type.

## Overall Subgroup
This subset of data explores observations when _Focus_ = I and _Focus_= They.

Read Data

```python
subgroup_theoretical = pd.read_csv("../../data/processed/subgroup_theoretical.csv")
```

1. Create escape_2

Binary Outcome: include responses that are not "L2_other": those who managed to escape L2_other by choosing either a correct response or an L1_transfer response.

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

2. Counts and Proportions

```python
escape_L2_counts = pd.crosstab(
    subgroup_theoretical["escape_L2"],
    subgroup_theoretical["Response_Full"])

escape_L2_counts
```

```python
escape_L2_props = subgroup_theoretical["escape_L2"].value_counts(normalize = True)
escape_L2_props
```

From 198 observations, almost 80% of responses managed to escape L2_other options. 

Counts and Proportions by Condition:

```python
counts_condition = pd.crosstab(
    [subgroup_theoretical["Goal_Type"], subgroup_theoretical["Agent"]],
    subgroup_theoretical["escape_L2"],
    margins = True)

counts_condition
```

```python
props_condition = pd.crosstab(
    [subgroup_theoretical["Goal_Type"], subgroup_theoretical["Agent"]],
    subgroup_theoretical["escape_L2"],
    normalize= "index")

props_condition
```

Agency appears to help participants escape L2. 

The condition with the strongest L1 transfer (goal_non_frequent, agent=1) also exhibits the highest L2 escape rate (95.8%). This suggests that failure does not occur at the L2 suppression stage, but at the subsequent resolution stage.

In addition, if there is a goal representation available, agency seems to help participants commit by leaving the L2 space; however, if there is no goal representation, agency seems to have much less to work with -which suggests that we could be dealing with an entire different behaviour when it comes to escaping l2 (and subsequent resolving) when having a goal representation available (*goal_frequent* and *goal_non_frequent*) vs. not having a goal representation available (*no_goal*).

3. Logistic Models

Import Libraries

```python
import statsmodels.formula.api as smf
from scipy.stats import chi2
```

```python
subgroup_theoretical["escape_L2"].value_counts()
```

**Model E1**

escape_L2 ~ Agent

```python
E1 = smf.logit(
    "escape_L2 ~ Agent",
    data=subgroup_theoretical
    ).fit()

print(E1.summary())
```

**Model E2**

escape_L2 ~ Goal_Type

```python
E2 = smf.logit(
    "escape_L2 ~ Goal_Type",
    data= subgroup_theoretical
    ).fit()

print(E2.summary())
```

```python
E3 = smf.logit(
    "escape_L2 ~ Agent + Goal_Type",
    data= subgroup_theoretical
    ).fit()

print(E3.summary())
```

**Model E4**

escape_L2 ~ Agent * Goal_Type

```python
E4 = smf.logit(
    "escape_L2 ~ Agent * Goal_Type",
    data=subgroup_theoretical
    ).fit()

print(E4.summary())
```

4. Model Comparisons

```python
from scipy.stats import chi2

def lr_test(model_small, model_large):

    LR = 2 * (model_large.llf - model_small.llf)

    df = model_large.df_model - model_small.df_model

    p = chi2.sf(LR, df)

    return pd.Series({
        "LR": LR,
        "df": df,
        "p_value": p
    })
```

```python
lr_test(E1, E3)
```

```python
lr_test(E1, E4)
```
