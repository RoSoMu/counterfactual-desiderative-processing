```python
subgroup_theoretical = pd.read_csv("../../data/processed/subgroup_theoretical.csv")
```

1. Create resolved_transfer

Binary Outcome: include L1_transfer responses from escape_L2 observations.

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

```python
escapees = subgroup_theoretical[subgroup_theoretical["escape_L2"] == 1].copy()
```

```python
escapees["resolved_transfer"] = (escapees["Response_Full"] == "L1_transfer").astype(int)
```

2. Counts and Proportions

```python
resolved_transfer_counts = pd.crosstab(
    escapees["resolved_transfer"],
    escapees["Response_Full"])

resolved_transfer_counts
```

```python
resolved_transfer_props = escapees["resolved_transfer"].value_counts(normalize = True)
resolved_transfer_props
```

From 160 observations, 60% of responses that had escaped *L2_other* options were resolved by restoring to L1_transfer. 

Counts and Proportions by Condition:

```python
counts_condition = pd.crosstab(
    [escapees["Goal_Type"], escapees["Agent"], escapees["Focus"]],
    escapees["resolved_transfer"],
    margins = True)

counts_condition
```

```python
props_condition = pd.crosstab(
    [escapees["Goal_Type"], escapees["Agent"], escapees["Focus"]],
    escapees["resolved_transfer"],
    normalize= "index")

props_condition
```

4. Model Comparisons

```python
from scipy.stats import chi2

def lr_test(model_small, model_large):

    LR = 2 * (model_large.llf - model_small.llf)

    df = model_large.df_model - model_small.df_model

    p = chi2.sf(LR, df)

    return pd.Series({
        "LR": LR,
        "df": df,
        "p_value": p
    })
```

```python
lr_test(RT01, RT03)
```

```python
lr_test(RT01, RT04)
```
