1. Create resolved_correct

Binary Outcome: correct responses from escape_L2 observations.

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

3. Logistic Models

Import Libraries

**Goals**

**Main Effect Models**

**RC01**

resolved_correct ~ Focus

```python
RC01 = smf.logit(
    "resolved_correct ~ Focus",
    data=goals_escapees
    ).fit()

print(RC01.summary())
```

Now that the data has been goal split, Focus gains more relevance. While still not significant and with fewer observations, Focus on its own as a model shows better than when the data is combined (LLR p-value: 0.2567).

**RC02**

resolved_correct ~ Agent

```python
RC02 = smf.logit(
    "resolved_correct ~ Agent",
    data=goals_escapees
    ).fit()

print(RC02.summary())
```

While the model with agency cannot predict anything, it is interesting how this preedictor, in contrast, does not seem to change at all compared to when the data is combined. This seems to confirm that Agency does not do much when it comes to resolving correctly. 

**RC03**

resolved_correct ~ Goal_Type

```python
RC03 = smf.logit(
    "resolved_correct ~ Goal_Type",
    data=goals_escapees
    ).fit()

print(RC03.summary())
```

The model with Goal Type improves compared to combined data although goal type can't still predict anything on its own. Goal_non_frequent's coefficient and p-value have remained untouched after pruning observations. 

**Additive Models**

**RC04**

resolved_correct ~ Focus + Agent

```python
RC04 = smf.logit(
    "resolved_correct ~ Focus + Agent",
    data=goals_escapees
    ).fit()

print(RC04.summary())
```

Again, Focus becomes more dominant when the data is split by goal in the additive model Focus-Agency (They p-value = 0.078) than when the data is combined (RC04 They p-value = 0.275)

**RC05**

resolved_correct ~ Focus + Goal_Type

```python
RC05 = smf.logit(
    "resolved_correct ~ Focus + Goal_Type",
    data=goals_escapees
    ).fit()

print(RC05.summary())
```

And once again, Focus becomes more dominant when the data is split by goal in the additive model Focus-Goal Type (They p-value = 0.100) than when the data is combined (RC05 They p-value = 0.252)

Focus alone's LLR p-value = 0.0866 and Goal Type alone's LLR p-value = 0.1404; together, LLR p-value = 0.848. Approaching significance, not there yet but way better than the same model with combined data (RC05 LLR p-value = 0.2303)

Goal_non_frequent's coefficient hardly changes compared to RC03 or the model with Goal_Type on its own (coeff = -0.586). The Focus coeffcient also sits around 0.69. So, both predictors seem to be behaving independently. 

**RC06**

resolved_correct ~  Agent + Goal_Type

```python
RC06 = smf.logit(
    "resolved_correct ~ Agent + Goal_Type",
    data=goals_escapees
    ).fit()

print(RC06.summary())
```

Goal_non_frequent's coefficient (-0.595) hardly changes compared to RCO5's (coeff = -0.570, with Focus) and RC03's (coeff = -0.586, Goal Type only). This shows that Goal Type could be the type of variable that contributes to explaining behaviour steadily without being individually significant.

4. Model Comparisons

```python
from scipy.stats import chi2

def lr_test(model_small, model_large):

    LR = 2 * (model_large.llf - model_small.llf)

    df = model_large.df_model - model_small.df_model

    p = chi2.sf(LR, df)

    return pd.Series({
        "LR": LR,
        "df": df,
        "p_value": p
    })
```

```python
lr_test(RC01, RC03)
```

```python
lr_test(RC01, RC04)
```
