Read Data

```python
subgroup_theoretical = pd.read_csv("../../data/processed/subgroup_theoretical.csv")
```

1. Create escape_2

Binary Outcome: responses that are not "L2_other": those who managed to escape L2_other by choosing either a *correct* response or an *L1_transfer* response.

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

Split data into goals vs. no_goals

```python
goals = subgroup_theoretical[subgroup_theoretical["Goal_Type"]!= "no_goal"].copy()
```

```python
no_goal = subgroup_theoretical[subgroup_theoretical["Goal_Type"]== "no_goal"].copy()
```

3. Logistic Models

Import Libraries

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
from scipy.stats import chi2
```

Sanity checks:

```python
goals["escape_L2"].value_counts()
```

```python
no_goal["escape_L2"].value_counts()
```

**Goals**

**Main Effect Models**

**E01**

escape_L2 ~ Focus

```python
E01 = smf.logit(
    "escape_L2 ~ Focus",
    data=goals
    ).fit()

print(E01.summary())
```

Focus on its own isn't enough to predict anything again -although it is way better on its own once there is a goal split in the data.

**E02**

escape_L2 ~ Agent

```python
E02 = smf.logit(
    "escape_L2 ~ Agent",
    data=goals
    ).fit()

print(E02.summary())
```

When the data was combined, agency on its own's p-value = 0.013 (coeff = 1.1086). Once the data is the goal split is introduced, agency coefficient increases (1.7327) and p-value improves ( 0.007). Thus, after removing 66 observations, the effect gets actually stronger. 

**E03**

escape_L2 ~ Goal_Type

```python
E03 = smf.logit(
    "escape_L2 ~ Goal_Type",
    data=goals
    ).fit()

print(E03.summary())
```

Goal Type still contributes nothing on its own. But, once again, working with the goal split data has improved the model's LLR p-value (0.1875) compared to combined data (0.4060)

**Additive Models**

**E04**

escape_L2 ~ Focus + Agent

```python
E04 = smf.logit(
    "escape_L2 ~  Focus + Agent",
    data=goals
    ).fit()

print(E04.summary())
```

Focus on its own couldn't predict anything. Again, once it is added to the model with agency, while it is still non-significant (p-value 0 0.080), it is much more informative than boefore. The coefficient also grows from 0.694 to 0.876; so, once agency is accounted for, focus explains additional variation in the data.

The whole model's LLR p-value (0.00135) is also better than agent alone (0.00163); not a lot but not requiring an interaction.

This model explains a lot of variation and demostrates parsimony.


**E05**

escape_L2 ~ Focus + Goal_Type

```python
E05 = smf.logit(
    "escape_L2 ~  Focus + Goal_Type",
    data=goals
    ).fit()

print(E05.summary())
```

Goal Type works better with Focus that on its own but the model is still very poor. 

**E06**

escape_L2 ~ Agent + Goal_Type

```python
E06= smf.logit(
    "escape_L2 ~  Agent + Goal_Type",
    data=goals
    ).fit()

print(E06.summary())
```

This is a good model even though the goal_non_frequent's p is still not siginificant. It is contributing to the model. Given this, a full additive model that was not considered in the beginning should be looked into.

**P01**

escape_L2 ~ Focus + Agent + Goal_Type

```python
P01= smf.logit(
    "escape_L2 ~  Focus + Agent + Goal_Type",
    data=goals
    ).fit()

print(P01.summary())
```

Goal Type's p-value improves a little bit and the coefficient increases. The whole model is better (LLR p-value: 0.001709 vs. 0.002758) although not better than the focus-agency additive model's p (0.001351). 

Yet, this model shows a better picture of how counterfactual processing is shaped by considering the 3 cognitive dimensions (Focus, Agency and Goal_Type). In this sense, Agency shows a strong contribution, Focus shows a consistent positive contribution approaching significance and Goal Type shows a modest positive contribution.

4. Model Comparisons

```python
from scipy.stats import chi2

def lr_test(model_small, model_large):

    LR = 2 * (model_large.llf - model_small.llf)

    df = model_large.df_model - model_small.df_model

    p = chi2.sf(LR, df)

    return pd.Series({
        "LR": LR,
        "df": df,
        "p_value": p
    })
```

```python
lr_test(E01, E03)
```

```python
lr_test(E01, E04)
```
