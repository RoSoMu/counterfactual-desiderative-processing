Read Data

```python
subgroup_theoretical = pd.read_csv("../../data/processed/subgroup_theoretical.csv")
```

1. Create resolved_transfer

Binary Outcome: transfer responses from those observations that managed to escape L2_other options.

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

```python
escapees = subgroup_theoretical[subgroup_theoretical["escape_L2"] == 1].copy()
```

```python
escapees["resolved_transfer"] = (escapees["Response_Full"] == "L1_transfer").astype(int)
```

Split escapees data into goals vs. no_goals

```python
goals_escapees = escapees[escapees["Goal_Type"] != "no_goal"].copy()
```

```python
no_goal_escapees = escapees[escapees["Goal_Type"] == "no_goal"].copy()
```

3. Logistic Models

Import Libraries

```python
import statsmodels.formula.api as smf
from scipy.stats import chi2
```

Sanity checks:

**Goals**

**Main Effect Models**

**RT01**

resolved_transfer ~ Focus

```python
RT01 = smf.logit(
    "resolved_transfer ~ Focus",
    data=goals_escapees
    ).fit()

print(RT01.summary())
```

With goal split dataset, the effect of Focus on its own seems to be more defined almost reaching significant.

**RT02**

resolved_transfer ~ Agent

```python
RT02 = smf.logit(
    "resolved_transfer ~ Agent",
    data=goals_escapees
    ).fit()

print(RT02.summary())
```

Agency still can't do anything on its own; however, directing the analysis to goal data has improved it a lot compared to combined data (RT02 LLR p_value = 0.9999).

**RT03**

resolved_transfer ~ Goal_Type

```python
RT03 = smf.logit(
    "resolved_transfer ~ Goal_Type",
    data=goals_escapees
    ).fit()

print(RT03.summary())
```

By eliminating no_goal, the model with Goal Type alone has improved a a lot. Goal_non_frequent keeps the same coefficient as with combined data (0.666) and the same p-value = 0.095.

**Additive Models**

**RT04**

resolved_transfer ~ Focus + Agent

```python
RT04 = smf.logit(
    "resolved_transfer ~ Focus + Agent",
    data=goals_escapees
    ).fit()

print(RT04.summary())
```

Just as with combined data, the additive model Focus-Agent cannot quite make it. Focus keeps p-value improves a little compared to the model with Focus on its own. The model has improved a lot compared to the same with model with combined data. 

**RT05**

resolved_transfer ~ Focus + Goal_Type

```python
RT05 = smf.logit(
    "resolved_transfer ~ Focus + Goal_Type",
    data=goals_escapees
    ).fit()

print(RT05.summary())
```

Even though the coefficients do not make it to significance, the model is significant overall. Goal_non_frequent's coefficient and p-values are almost the same as the model with combined data; it is Focus what that has improved once the no_goal variation was removed. This is interesting as it does give way to consider that, despite not having enough power, Goal Type could after all play a role. 

**RT06**

resolved_transfer ~ Agent + Goal_Type

```python
RT06 = smf.logit(
    "resolved_transfer ~ Agent + Goal_Type",
    data=goals_escapees
    ).fit()

print(RT06.summary())
```

This model also improves compared to the combined dataset's. And, although goal_non_frequent's p-value doesn't reach significance (but it almost matches the combined data's at 0.095), it does show how it stands compared to Agency when it comes to resolving with transfer. 

**Interactive Models**

**RT07**

resolved_transfer ~ Focus * Agent

```python
RT07 = smf.logit(
    "resolved_transfer ~ Focus * Agent",
    data=goals_escapees
    ).fit()

print(RT07.summary())
```

Compaired to the combined dataset, Focus p-value improves a little bit and so does the Agent's. The interaction's p-values increases this time while keeping the same coefficient, with the same magnitude and the same direction. This could just be as a result of reducing observations and not having enough power. The model's overall LLR p-value also improves compared to combined data although very slightly. 

**RT08**

resolved_transfer ~ Focus * Goal_Type

```python
RT08 = smf.logit(
    "resolved_transfer ~ Focus * Goal_Type",
    data=goals_escapees
    ).fit()

print(RT08.summary())
```

**RT09**

resolved_transfer ~ Agent * Goal_Type

```python
RT09 = smf.logit(
    "resolved_transfer ~ Agent * Goal_Type",
    data=goals_escapees
    ).fit()

print(RT09.summary())
```

**RT10**

resolved_transfer ~ Focus * Agent + Goal_Type

```python
RT10 = smf.logit(
    "resolved_transfer ~ Focus * Agent + Goal_Type",
    data=goals_escapees
    ).fit()

print(RT10.summary())
```

For the full model on goal split data, Focus and Agent keep similar significance and the interaction's coefficient, while not significant, is similar to the previous one and the same direction. 
Goal type's contribution remains very steady, with a very similar contribution (coeff = 0.6518) compared to the additive model Focus-Goal_Type RT05 (coeff = 0.6531) and p-value = 0.115 (vs. 0.108)

4. Model Comparisons

```python
from scipy.stats import chi2

def lr_test(model_small, model_large):

    LR = 2 * (model_large.llf - model_small.llf)

    df = model_large.df_model - model_small.df_model

    p = chi2.sf(LR, df)

    return pd.Series({
        "LR": LR,
        "df": df,
        "p_value": p
    })
```

```python
lr_test(RT02, RT03)
```

```python
lr_test(RT02, RT04)
```
