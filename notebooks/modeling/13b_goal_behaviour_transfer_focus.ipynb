Read Data

```python
subgroup_theoretical = pd.read_csv("../../data/processed/subgroup_theoretical.csv")
```

1. Create resolved_transfer

Binary Outcome: transfer responses from those observations that managed to escape L2_other options.

```python
subgroup_theoretical["escape_L2"] = (
    subgroup_theoretical["Response_Full"] != "L2_other"
).astype(int)
```

```python
escapees = subgroup_theoretical[subgroup_theoretical["escape_L2"] == 1].copy()
```

```python
escapees["resolved_transfer"] = (escapees["Response_Full"] == "L1_transfer").astype(int)
```

Split escapees data into goals vs. no_goals

```python
goals_escapees = escapees[escapees["Goal_Type"] != "no_goal"].copy()
```

```python
no_goal_escapees = escapees[escapees["Goal_Type"] == "no_goal"].copy()
```

From 106 observations, 41.5% of goal responses that had escaped *L2_other* options were resolved by means of transfer. Slightly higher than resolving correctly (40.5%) but not that much. 

resolved_correct responses proportions for no_goals:

```python
no_goal_resolved_transfer_props = no_goal_escapees["resolved_transfer"].value_counts(normalize=True)
no_goal_resolved_transfer_props
```

From 54 observations, 37% of no_goal responses that had escaped *L2_other* options were resolved by means of transfer. Also higher than resolving correctly (33.3%) but not that much either.

- The proportion distribution of transfer resolution responses is higher when there is a goal oriented representation (41.5%) as opposed to when there isn't a representation of a goal (37%). 
- However, the gap narrows compared to the distribution of correct resolution responses between goal oriented vs. no goal representations.

4. Model Comparisons

```python
from scipy.stats import chi2

def lr_test(model_small, model_large):

    LR = 2 * (model_large.llf - model_small.llf)

    df = model_large.df_model - model_small.df_model

    p = chi2.sf(LR, df)

    return pd.Series({
        "LR": LR,
        "df": df,
        "p_value": p
    })
```

```python
lr_test(RT02, RT03)
```

```python
lr_test(RT02, RT04)
```
